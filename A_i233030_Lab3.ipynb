```python

paths = {
    "W": {"distance": 300, "time": 25, "congestion": "Medium"},
    "X": {"distance": 200, "time": 15, "congestion": "Low"},
    "Y": {"distance": 150, "time": 20, "congestion": "Medium"},
    "Z": {"distance": 250, "time": 10, "congestion": "High"}
}

congestion_values = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}

def utility(path):
    distance = path["distance"]
    time = path["time"]
    congestion = congestion_values[path["congestion"]]

    score = (0.4 * distance + 0.4 * time + 0.2 * congestion)
    return score

utility_scores = {}

for path_name, attributes in paths.items():
    utility_scores[path_name] = utility(attributes)

best_path = max(utility_scores, key=utility_scores.get)

print("Utility Scores for Each Path:\n")

for path, score in utility_scores.items():
    print(f"Path {path}: Utility = {score:.2f}")

print("\nSelected Best Path:", best_path)
```

# Task 3

```python
import random


class Greenhouse:
    def __init__(self):
        self.state = {
            "Section 1": random.choice(["Healthy", "Wilted"]),
            "Section 2": random.choice(["Healthy", "Wilted"])
        }

    def get_observation(self, section):
        """Agent can only observe current section"""
        return self.state[section]

    def water(self, section):
        """Watering makes wilted plants healthy"""
        self.state[section] = "Healthy"



class PlantCareAgent:
    def __init__(self):
        self.internal_state = {
            "Section 1": "Unknown",
            "Section 2": "Unknown"
        }
        self.location = "Section 1"

    def perceive_and_update(self, observation):
        """Update internal memory based on observation"""
        self.internal_state[self.location] = observation

    def decide_action(self):
        """Choose action based on internal model"""
        current_condition = self.internal_state[self.location]

        if current_condition == "Wilted":
            return "Water"
        else:
            return "Move"

    def move(self):
        """Move agent to other section"""
        self.location = "Section 2" if self.location == "Section 1" else "Section 1"

def simulate(time_steps=6):
    env = Greenhouse()
    agent = PlantCareAgent()

    print("\n GREENHOUSE MONITORING SIMULATION START\n")

    for step in range(1, time_steps + 1):

        print(f"--- Time Step {step} ---")

        print("REAL ENVIRONMENT STATE:", env.state)
        print("AGENT INTERNAL STATE  :", agent.internal_state)

        observation = env.get_observation(agent.location)
        print(f"\nAgent observes {agent.location}: {observation}")

        agent.perceive_and_update(observation)

        action = agent.decide_action()

        if action == "Water":
            print(f"Action: Watering plants in {agent.location}")
            env.water(agent.location)
        else:
            print("Action: Moving to other section")
            agent.move()

        print("\nUpdated Agent Internal State:", agent.internal_state)
        print("Agent Current Location:", agent.location)
        print("\n")

    print(" SIMULATION COMPLETE\n")

simulate()
```

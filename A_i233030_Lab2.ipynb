# Task 2

```python
import seaborn as sns
import pandas as pd
df_tips = sns.load_dataset("titanic")
targetVar = 'embark_town'
numericalFeature = df_tips.select_dtypes(include= ['number']).columns.tolist()
categoricalFeature = df_tips.select_dtypes(exclude=['number']).columns.tolist()
roles = []
for i in df_tips.columns:
    if i == targetVar:
        roles.append({'Attribute': i, 'Role': 'Target Variable'})
    elif i in numericalFeature:
        roles.append({'Attribute': i, 'Role': 'Numerical Feature'})
    else:
        roles.append({'Attribute': i, 'Role': 'Categorical Feature'})
print(pd.DataFrame(roles))
```

```python
import seaborn as sns
df_titanic = sns.load_dataset("titanic")
df_titanic.isnull().sum()
#imputation of age
df_titanic["age"] = df_titanic["age"].fillna(df_titanic["age"].mean())
#imputation of embarked column
df_titanic["embarked"] = df_titanic["embarked"].fillna(df_titanic["embarked"].mode())
#imputation of deck column
df_titanic.drop(columns=['deck'], inplace = True)
df_titanic.isnull().sum()
```

# Task 7

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_tips = sns.load_dataset("tips")
billQ1 = df_tips['total_bill'].quantile(0.25)
billQ3 = df_tips['total_bill'].quantile(0.75)
billIQR = billQ1 - billQ3
billlower_bound = billQ1 - 1.5 * billIQR
billupper_bound = billQ3 + 1.5 * billIQR
df_tips.boxplot(column=['total_bill'])
tipQ1 = df_tips['tip'].quantile(0.25)
tipQ3 = df_tips['tip'].quantile(0.75)
tipIQR = tipQ1 - tipQ3
tiplower_bound = tipQ1 - 1.5 * tipIQR
tipupper_bound = tipQ3 + 1.5 * tipIQR
df_tips.boxplot(column=['tip'])
outliers = df_tips[(df_tips['total_bill'] < billlower_bound) | 
(df_tips ['total_bill'] > billupper_bound)]
print(outliers)
tipoutliers = df_tips[(df_tips['tip'] < tiplower_bound) | 
(df_tips ['tip'] > tipupper_bound)]
print(tipoutliers)
```

```python
myList = [10,20,3,4,23,43]
evenNums=list(filter(lambda x:x%2==0, myList))
evenResult=list(map(lambda x:x**3, evenNums))
evenResult
oddNums=list(filter(lambda x:x%2!=0, myList))
oddResult=list(map(lambda x:x**2, oddNums))
oddResult
Result = evenResult+oddResult
SortedResult = sorted(Result)
SortedResult
```

# Task 2

```python
Products = {}

for i in range(0, 6):
    name = input("Enter the product name: ") #user input
    cat = input("Enter product category: ")
    price = float(input("Enter price: "))

    Products[i] = {
        "Prod_Name": name,
        "Prod_Category": cat,
        "price": price
    }
    minProduct = min(Products.items(), key=lambda item: item[1]["price"])
    print(minProduct)
    maxProduct = max(Products.items(), key=lambda item: item[1]["price"])
    print(maxProduct)
    sorted_products = sorted(Products.items(), key=lambda item: item[1]["price"])
    print(sorted_products)
```

# Task 3

```python
import functools
List = [1,2,3,4,45,3,12,34,23,54]
evenNums=list(filter(lambda x:x%2==0, List))
evenResult=list(map(lambda x:x**2, evenNums))
prod = functools.reduce(lambda x,y: x+y, evenResult)
print(prod)
```

# Task 7

```python
import numpy as np
random_matrix = np.random.rand(100, 3)
Colmin = np.max(random_matrix, axis=0)
Colmax = np.min(random_matrix, axis=0)
normalized_Data = (random_matrix - Colmin) / (Colmax - Colmin)
variances = normalized_Data.var(axis=0)
highestVarColIndex = np.argmax(variances)
print(highestVarColIndex)
```
